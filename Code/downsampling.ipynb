```python
import os
import sys
import json
import numpy as np
import pandas as pd
from sklearn import preprocessing
from PIL import Image
import matplotlib.pyplot as plt
```

```python
Dataset_path = "../Dataset/images"
```

```python
images = os.listdir(Dataset_path)
print(len(images))
```

```python
# Load the first image
image_1_path = os.path.join(Dataset_path, images[0])
image_1 = Image.open(image_1_path)

# View the image
image_1_np = np.array(image_1)

# Get the shape of the image
print(image_1_np.shape)
# plot the image
plt.imshow(image_1_np)
plt.show()
```

```python
# downsample the image to 512x512
image_1 = image_1.resize((128, 128))
image_1_np = np.array(image_1)
print(image_1_np.shape)
plt.imshow(image_1_np)
plt.show()
```

```python
# loop over the images and downsample them to 128 x 128 and save them in new folder
downsampled_images_path = "../Dataset/downsampled_images"
i = 0
for image in images:
    image_name =  image.split(".")[0] 
    image_path = os.path.join(Dataset_path, image)
    image = Image.open(image_path)
    image = image.resize((128, 128))
    # save as numpy array
    image_np = np.array(image)
    # save as numpy array
    np.save(os.path.join(downsampled_images_path, image_name), image_np)
    i += 1
    print(i)

    
```

```python
# load the downsampled image
downsampled_images = os.listdir(downsampled_images_path)
image_1_path = os.path.join(downsampled_images_path, downsampled_images[6])
image_1 = np.load(image_1_path )
print(image_1.shape)
```
